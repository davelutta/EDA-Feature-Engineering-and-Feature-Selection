```python
def elapsed_years(df, var):
    df[var] =df['YrSold'] - df[var]
    return df
```

```python
for var in submission_vars:
    if x_train[var].dtype == 'O':  # Check if the variable is categorical (object type)
        # Fill missing values with the mode (most frequent value) from X_train
        mode_value = x_train[var].mode()[0]  # Mode might return multiple values; [0] gives the most frequent one
        test[var].fillna(mode_value, inplace=True)
    else:
        # For numeric variables, fill missing values with the median from X_train
        median_value = x_train[var].median()
        test[var].fillna(median_value, inplace=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score
import numpy as np

def tree_binariser(var, x_train, y_train, x_test, train):
    score_ls = []  # Changed score_Is to score_ls for consistency
    for tree_depth in [1, 2, 3, 4]:
        tree_model = DecisionTreeRegressor(max_depth=tree_depth)
        scores = cross_val_score(tree_model, x_train[var].to_frame(), y_train, cv=3, scoring='neg_mean_squared_error')
        score_ls.append(np.mean(scores))
    depth = [1, 2, 3, 4][np.argmin(score_ls)]
    # print(score_ls, np.argmin(score_ls), depth)
    tree_model = DecisionTreeRegressor(max_depth=depth)
    tree_model.fit(x_train[var].to_frame(), y_train)
    
    x_train[var] = tree_model.predict(x_train[var].to_frame())
    x_test[var] = tree_model.predict(x_test[var].to_frame())
    train[var] = tree_model.predict(train[var].to_frame())

# Example usage (assuming x_train, y_train, x_test, and train are already defined)
# tree_binariser('your_variable_name', x_train, y_train, x_test, train)

    

```

```python
for var in numerical:
    tree_binariser(var, x_train, y_train, x_test, test)
```

```python
def rare_imputation(variable):
    temp = x_train.groupby([variable])[variable].count()/ float(len(x_train))
    frequent_cat = [x for x in temp.loc[temp>0.03].index.values]
    x_train[variable] = np.where(x_train[variable].isin(frequent_cat), x_train[variable], 'Rare')
    x_test[variable] = np.where(x_test[variable].isin(frequent_cat), x_test[variable], 'Rare')
    test[variable] = np.where(test[variable].isin(frequent_cat), test[variable], 'Rare')
```

```python
for var in ['BsmtFullBath', 'BsmtHalfBath', 'GarageCars']:
    test[var] = test[var].fillna(0).astype('int')  
```

```python
# find infrequent labels in categorical variables and replace by Rare
for var in categorical:
    rare_imputation(var)
```

```python
import pandas as pd
import numpy as np




# Fill missing values with median (only for numeric columns)
for var in x_train.columns:
    if var != 'SalePrice' and test[var].dtype in ['int64', 'float64']:  # Check if column is numeric
        try:
            # Convert column to numeric if possible
            test[var] = pd.to_numeric(test[var], errors='raise')  # errors='raise' will raise an error if conversion fails
            if test[var].isnull().any():
                median_val = test[var].median()  # Calculate median of the column
                test[var].fillna(median_val, inplace=True)  # Fill missing values with the median
                submission_vars.append(var)
            else:
                print(f"No missing values found in {var}")
        except ValueError as e:
            print(f"Skipping {var} due to non-numeric data: {e}")

print("Test DataFrame after handling missing values:")
print(test)
print("submission_vars:", submission_vars)
```

```python
def encode_categorical_variable(var, target):
    ordered_labels = x_train.groupby([var])[target].mean().to_dict()
    x_train[var] = x_train[var].map(ordered_labels)
    x_test[var] = x_test[var].map(ordered_labels)
    test[var] = test[var].map(ordered_labels)
    
    
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
from math import sqrt

# Assuming x_train, y_train, x_test, y_test are defined DataFrames/Series

# Define training and evaluation sets
eval_set = [(x_test[training_vars], np.log(y_test))]

# Instantiate XGBoost model
xgb_model = xgb.XGBRegressor()
xgb_model.fit(x_train[training_vars], np.log(y_train), eval_set=eval_set, verbose=False)

# Predictions on training set
pred_train = xgb_model.predict(x_train[training_vars])
print('xgb train mse: {}'.format(mean_squared_error(y_train, np.exp(pred_train))))
print('xgb train rmse: {}'.format(sqrt(mean_squared_error(y_train, np.exp(pred_train)))))
print()

# Predictions on test set
pred_test = xgb_model.predict(x_test[training_vars])
print('xgb test mse: {}'.format(mean_squared_error(y_test, np.exp(pred_test))))
print('xgb test rmse: {}'.format(sqrt(mean_squared_error(y_test, np.exp(pred_test)))))
```
